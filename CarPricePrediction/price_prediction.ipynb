## 2.13 Regularization


```python
X = [[4, 4, 4], [3, 5, 5], [5, 1, 1], [5, 4, 4], [7, 5, 5], [4, 5, 5.00000001]]

X = np.array(X)
X
```

```python
y = [1, 2, 3, 1, 2, 3]
```

```python
XTX = X.T.dot(X)
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)  # LinAlgError: Singular matrix
```

```python
XTX_inv.dot(X.T).dot(y)
```

Adding small number to diagonal of matrix, to make it invertible.


```python
XTX = [[1.0001, 2, 2], [2, 1.0001, 1.0000001], [2, 1.0000001, 1.0001]]

XTX = np.array(XTX)
```

```python
np.linalg.inv(XTX)
```

```python
XTX = XTX + 0.01 * np.eye(3)
```

```python
np.linalg.inv(XTX)
```

Regularization means controlling the weights that they don't grow too much.


```python
def train_linear_regression_reg(X, y, r=0.001):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])

    XTX = X.T.dot(X)
    XTX = XTX + r * np.eye(XTX.shape[0])

    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)

    return w_full[0], w_full[1:]
```

```python
X_train = prepare_X(df_train)
w0, w = train_linear_regression_reg(X_train, y_train, r=0.01)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)

rmse(y_val, y_pred)
```

## 2.14 Tuning the model


Finding the best parameter `r` for the model.


```python
for r in [0, 0.00001, 0.0001, 0.001, 0.1, 1, 10]:
    X_train = prepare_X(df_train)
    w0, w = train_linear_regression_reg(X_train, y_train, r=r)

    X_val = prepare_X(df_val)
    y_pred = w0 + X_val.dot(w)
    score = rmse(y_val, y_pred)

    print(f"r: {r:>6} - w0: {w0:<22} - score: {score}")
```

```python
r = 0.001
X_train = prepare_X(df_train)
w0, w = train_linear_regression_reg(X_train, y_train, r=r)

X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)
score = rmse(y_val, y_pred)
score
```

Der Fehler "LinAlgError: Singular matrix" tritt auf, wenn versucht wird, eine singuläre Matrix zu invertieren. Eine singuläre Matrix ist eine quadratische Matrix, die nicht invertierbar ist24.

Dieser Fehler entsteht in der Regel aus folgenden Gründen:

1. Die Matrix hat keinen vollen Rang, d.h. ihre Spalten oder Zeilen sind linear abhängig.

2. Die Determinante der Matrix ist Null.

3. Es gibt Multikollinearität zwischen den Variablen im Datensatz.

Um dieses Problem zu beheben, können Sie folgende Schritte in Betracht ziehen:

1. Daten überprüfen: Untersuchen Sie Ihre Eingabedaten auf mögliche Probleme wie fehlende Werte oder stark korrelierte Variablen.

2. Variablen entfernen: Entfernen Sie Variablen, die stark mit anderen korrelieren oder wenig Varianz aufweisen.

3. Regularisierung: Verwenden Sie Regularisierungstechniken wie Ridge-Regression oder Lasso, um die Stabilität zu erhöhen.

4. Pseudoinverse: Anstatt die inverse Matrix zu berechnen, können Sie die Pseudoinverse (Moore-Penrose-Inverse) verwenden.

5. Dimensionsreduktion: Techniken wie Hauptkomponentenanalyse (PCA) können helfen, die Dimensionalität zu reduzieren und Multikollinearität zu beseitigen.

6. Numerische Stabilität: Überprüfen Sie die numerische Stabilität Ihrer Berechnungen und erwägen Sie die Verwendung von Techniken zur Verbesserung der Konditionszahl der Matrix.

In Ihrem spezifischen Fall, wo Sie die CoxPHFitter-Klasse verwenden, könnte es hilfreich sein, den Penalizer-Wert zu erhöhen oder problematische Variablen zu identifizieren und zu entfernen2.
